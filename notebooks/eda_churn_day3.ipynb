```python
df['Total Charges'].dtype
```

```python
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors='coerce')
```

```python
df.isnull().sum()
```

```python
# DROP rows with missing Total Charges

df = df.dropna(subset=['Total Charges'])

# Verify no more missing values
df.isnull().sum()
```

```python
df.shape
```

```python
# Percentage of Churned Customers
churn_distribution = df['Churn Label'].value_counts(normalize=True) * 100
churn_distribution
```

```python
# churn by contract
churn_by_contract = df.groupby('Contract')['Churn Label'].value_counts(normalize=True).unstack() * 100
churn_by_contract
```

```python
# Avg monthly charges by churn vs retained
avg_monthly_charges = df.groupby('Churn Label')['Monthly Charges'].mean()
avg_monthly_charges
```

```python
import matplotlib.pyplot as plt

avg_monthly_charges.plot(kind='bar', color=['green', 'red'])
plt.title('Average Monthly Charges by Churn Status')
plt.ylabel('Average Monthly Charges')
plt.xlabel('Churn Status')
plt.show()
```

1. Two years contracts have the lowest churn rate, while month-to-month contracts have the highest churn rate. This suggests that customers with longer-term commitments are less likely to churn.
2. Churned chargers have a higher average monthly charge compared to retained customers, indicating that higher charges may contribute to customer churn.
3. the churn rate is approximately 26.5%, meaning that about one in four customers are leaving the service.

```python
# churn by payment method
churn_by_payment = df.groupby('Payment Method')['Churn Label'].value_counts(normalize=True).unstack() *100
churn_by_payment
```

```python
# Churn by internet service
churn_by_internet = df.groupby('Internet Service')['Churn Label'].value_counts(normalize=True).unstack() * 100
churn_by_internet
```

1. Electronic check payment method has the highest churn rate among all payment methods, suggesting that customers using this method may be more likely to leave the service.
2. Electronic check has a 45% churn rate, significantly higher than other payment methods
3.Customers with Fiber optic internet service have a higher churn rate compared to those with DSL or No internet service, indicating that the type of internet service may influence customer retention.
4. Cumtomers with No internet service have the lowest churn rate at around 8%, while those with Fiber optic service have the highest churn rate at approximately 40%.
